```python
import pandas as pd
import statistics

df = pd.read_csv('/content/bhp (1).csv')

#Remove outliers using percentile technique first. Use [0.001, 0.999] for lower and upper bound percentiles

upper_limit_new = df['price_per_sqft'].quantile(0.999)
lower_limit_new = df['price_per_sqft'].quantile(0.001)

print("Upper Limit: ", upper_limit_new)
print("Lower Limit: ", lower_limit_new)

#New dataframe 'no_outliers' excluding the percentile ouliers below 0.001 and above 0.999

no_outliers = df[(df['price_per_sqft'] < upper_limit_new) & (df['price_per_sqft'] > lower_limit_new)]

no_outliers
```

```python
#On the 'no_outliers' dataframe, use 4 standard deviation to remove outliers

#clarify the minimum and maximum limits
min_limit = no_outliers['price_per_sqft'].mean() - 4 * no_outliers['price_per_sqft'].std()
max_limit = no_outliers['price_per_sqft'].mean() + 4 * no_outliers['price_per_sqft'].std()

max_limit, min_limit

no_outliers[(no_outliers['price_per_sqft'] < max_limit) & (no_outliers['price_per_sqft'] > min_limit)]

```

```python
#Use zscore of 4 to remove outliers. This is quite similar to step (3) and you will get exact same result

no_outliers['zscore'] = (no_outliers.price_per_sqft - no_outliers.price_per_sqft.mean())/no_outliers.price_per_sqft.std()

no_outliers[(no_outliers['zscore'] < 4) & (no_outliers['zscore'] > -4)]
```
